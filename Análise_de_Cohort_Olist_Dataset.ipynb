```python
olist = df[['order_id','order_purchase_timestamp', 'seller_id']].drop_duplicates()
```

```python
olist
```

```python
olist['order_purchase_timestamp'] = pd.to_datetime(olist['order_purchase_timestamp'], format="%Y-%m-%d %H:%M:%S")
```

```python
olist
```

```python
olist['order_purchase_timestamp'] = olist['order_purchase_timestamp'].dt.to_period('M')
```

```python
olist
```

```python
olist['Cohort'] = olist.groupby('seller_id')['order_purchase_timestamp'].transform('min')
```

```python
olist
```

```python
olist_cohort = olist.groupby(['Cohort', 'order_purchase_timestamp']).agg(n_customers=('seller_id', 'nunique')).reset_index(drop=False)
olist_cohort
```

```python
olist_cohort['Period'] = (olist_cohort.order_purchase_timestamp - olist_cohort.Cohort).apply(lambda x:x.n)
```

```python
olist_cohort
```

```python
cohort_pivot = olist_cohort.pivot_table(index='Cohort', columns='Period',values='n_customers')
cohort_pivot
```

```python
cohort_size = cohort_pivot.iloc[:,0]
cohort_size
```

```python
retention_matrix = cohort_pivot.divide(cohort_size,axis=0)
retention_matrix
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# plotando
_ = plt.subplots(figsize=(12,8))
_ = sns.heatmap(
data= retention_matrix,
mask = retention_matrix.isnull(),
annot = True,
fmt = '.0%',
cmap = 'RdYlGn'
)
_ = plt.title('Matriz de Retenção', size = 14)
```
